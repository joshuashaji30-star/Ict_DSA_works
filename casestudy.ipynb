```python
balance= 1000


def deposit(balance):
    amount = int(input("Enter amount to deposit: "))
    balance = balance + amount
    print("Deposited amount:", amount)
    print("\n")
    return balance


def withdraw(balance):
    amount = int(input("Enter amount to withdraw: "))
    if amount <= balance:
        balance = balance - amount
        print("Withdrawn amount:", amount)
        print("\n")
    else:
        print("Insufficient balance")
    return balance

def check_balance(balance):
    print("\n")
    print("Avaialable Balance:", balance)


balance = deposit(balance)
balance = withdraw(balance)
check_balance(balance)
```

```python
t1 = float(input("Enter temperature 1: "))
t2 = float(input("Enter temperature 2: "))

h1 = float(input("Enter humidity 1: "))
h2 = float(input("Enter humidity 2: "))

def weather(t1, t2, h1, h2):
    avg_temp = (t1 + t2) / 2
    avg_hum = (h1 + h2) / 2

    print("Average Temperature:", avg_temp)
    print("Average Humidity:", avg_hum)

    if avg_temp > 30 and avg_hum < 50:
        print("Weather: Sunny")
    elif avg_hum > 70:
        print("Weather: Rainy")
    else:
        print("Weather: Cloudy")

weather(t1, t2, h1, h2)
```
