#Load SpaceX Launch Data From API

```python
import requests
import pandas as pd
```

```python


url = "https://api.spacexdata.com/v4/launches"
data = requests.get(url).json()

df = pd.DataFrame(data)
df = df[['name', 'date_utc', 'success', 'details', 'rocket']]
df
```

```python
df['date_utc'] = pd.to_datetime(df['date_utc'])
df['year'] =df['date_utc'].dt.year
df[['date_utc','year']]
```

#Load Rocket Metadata

```python
url="https://api.spacexdata.com/v4/rockets"
data=requests.get(url).json()
```

```python
r_df=pd.DataFrame(data)
```

```python
r_df = r_df[['id', 'name', 'type', 'active', 'stages']]
r_df
```

#Merge Launch and Rocket data

```python
merged_df = pd.merge(
    df,
    r_df,
    left_on='rocket',
    right_on='id',
    how='left'
)

merged_df.rename(columns={
    'name_x': 'launch_name',
    'name_y': 'rocket_name'
}, inplace=True)

merged_df.drop(columns=['id'], inplace=True)
merged_df
```

#Add Simulated Country Column

```python
import random
```

```python
countries = ['USA', 'Russia', 'India', 'China', 'France']

merged_df['country'] = [random.choice(countries) for _ in range(len(merged_df))]
merged_df[['launch_name','country']]
```

#Store Merged data in sqlite3

```python
import sqlite3
```

```python
conn = sqlite3.connect("spacex.db")

merged_df.to_sql("launches", conn, if_exists="replace", index=False)
```

#SQL Analysis

##Launches by country

```python
q1 = """
SELECT country, COUNT(*) AS All_launches
FROM launches
GROUP BY country
ORDER BY All_launches DESC;
"""
print(pd.read_sql(q1, conn))
```

##Year with highest launch

```python
q2 = """
SELECT year, COUNT(*) AS All_launches
FROM launches
GROUP BY year
ORDER BY All_launches DESC
LIMIT 1;
"""
print(pd.read_sql(q2, conn))
```

##Top 5 missions by Launch count

```python
q3 = """
SELECT launch_name, COUNT(*) AS launch_count
FROM launches
GROUP BY launch_name
ORDER BY launch_count DESC
LIMIT 5;
"""
print(pd.read_sql(q3, conn))
```
